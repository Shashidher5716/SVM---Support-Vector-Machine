```python
# Importing the Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the Test dataset

df1 = pd.read_csv('SalaryData_Test.csv')
```

```python
# Checking the head of Test dataset

df1.head()
```

```python
# Loading the Train dataset

df2 = pd.read_csv('SalaryData_Train.csv')
```

```python
# Checking the head of Train dataset

df2.head()
```

```python
# Importing 1hot Encoder

from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
df1['education'] = enc.fit_transform(df1['education'])
df1['workclass'] = enc.fit_transform(df1['workclass'])
df1['maritalstatus'] = enc.fit_transform(df1['maritalstatus'])
df1['occupation'] = enc.fit_transform(df1['occupation'])
df1['relationship'] = enc.fit_transform(df1['relationship'])
df1['race'] = enc.fit_transform(df1['race'])
df1['sex'] = enc.fit_transform(df1['sex' ])
df1['native'] = enc.fit_transform(df1['native'])
```

```python
df1
```

```python
df1['Salary'].value_counts()
```

```python
df1['Salary'] = enc.fit_transform(df1['Salary'])
```

```python
df1
```

```python
# Standardization using Scaler Min = 0, Max = 1

def norm_fun(i):
  x = (i - i.min())/(i.max() - i.min())

  return x
```

```python
data_norm = norm_fun(df1)
data = pd.DataFrame(data_norm)
data.describe()
```

```python
# Checking the correalation

df1.corr()
```

```python
# Heat map

fig = plt.figure(figsize = (22,15))
sns.heatmap(df1.corr(), cmap='viridis', annot = True) 
```

```python
# Spliting the data

X = df1.iloc[:,0:13]
Y = df1.iloc[:,13]
```

```python
# Importing the train test split

from sklearn.model_selection import train_test_split

x_train, x_test , y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
# Building the Model 

from sklearn.svm import SVC
```

```python
model_linear = SVC(kernel = 'linear')
model_linear.fit(x_train, y_train)
y_pred_linear = model_linear.predict(x_test)

np.mean(y_test == y_pred_linear)
```

```python
model_poly = SVC(kernel = 'poly')
model_poly.fit(x_train, y_train)
y_pred_poly = model_linear.predict(x_test)

np.mean(y_test== y_pred_linear)
```

```python
model_rbf = SVC(kernel = 'rbf')
model_rbf.fit(x_train, y_train)
y_pred_rbf = model_linear.predict(x_test)

np.mean(y_test== y_pred_linear)
```

```python
model_sigmoid = SVC(kernel = 'sigmoid')
model_sigmoid.fit(x_train, y_train)
y_pred_sigmoid = model_linear.predict(x_test)

np.mean(y_test== y_pred_linear)
```
